```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[465.01235416], [441.60448968], [488.3920215 ], [449.65922921], [471.90306405]])
y = np.array([[463.26], [444.37], [488.56], [446.48], [473.9 ]])


plt.figure(figsize=(8, 4))  # figsize:确定画布大小 

plt.scatter(x, y, c='red', label='true value - predicted value') 
plt.plot(x, x, 'r', label='y=x')
plt.xlabel(u'true value', fontsize=18)
plt.ylabel(u'training set', fontsize=18)

plt.legend()  # 显示图例

plt.show()  # 显示所绘图形
```
